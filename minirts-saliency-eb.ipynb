## Load a pretrained model

```python
class Args():
    pass
args = Args()
args.num_actions = 4
args.save_dir = './saved/'

state_to_torch = lambda state: Variable(torch.Tensor(state)).view(1,*state.shape) # add a batch dimension

model = NNPolicy(num_actions=args.num_actions)
ckpt_reward = model.try_load(args.save_dir)
print('this saved model had a mean reward of {:.2f}'.format(ckpt_reward))
```

## Run the agent through the environment

**Reward landscape.** Also, let's review the reward landscape. When the agent chooses a tower to attack, its reward is computed as follows:
 * `if tower_health < agent_health`
     * `reward = tower_health`
 * `else`
     * `reward = -1 # this is a relatively large negative reward`

Based on this reward landscape, the expected value of an attack, which is a good proxy for the agent's preference to attack a tower might look like this:
![attack preference](static/attack-preference.png)


```python
env = Towers()
N = 5000
rewards = np.zeros((N))

for i in range(N):
    obs = env.reset()
    state = state_to_torch(obs)
    value, probs = model(state)
    m = torch.distributions.Categorical(probs)
    action =  m.probs.max(1)[1].data.view(-1)[0] # choose the action that has MAX probability
    rewards[i] = env.step(action)[1]
    
print('reward mean: {:.4f}, reward std: {:.4f}'.format(rewards.mean(), rewards.std()))
```

## Enter EB Mode
Refer to **Alg. 1** in the [original paper](https://arxiv.org/abs/1608.00507). The relevant (general) equation is:

$$ P_n = A_n \odot W^+ \left(\frac{P_{n-1}}{W^{+T}A_n}\right) $$
    
We need to do this to the Linear and Conv functions for the model we just trained. Other functions that need this extension include:
 * average pooling
 * any pointwise function that has a min value less than 0
     * these must have their min values subtracted from their outputs

```python
eb.use_eb(True)
```

```python
# reproducibility
seed = 2 ; np.random.seed(seed) ; torch.manual_seed(seed)

# get an input
obs = env.reset() + np.random.randn(*obs.shape)*0.05
inputs = state_to_torch(obs)

# perform excitation backprop to get saliency map
dist = model(inputs)[1]
np_dist = dist.data.numpy().ravel() # compute distribution
most_prob_action_ix = np.argmax(np_dist)
prob_outputs = Variable(torch.zeros(dist.shape)) ; prob_outputs.data[:,most_prob_action_ix] += 1
prob_inputs = eb.excitation_backprop(model, inputs, prob_outputs, contrastive=True)

saliency = prob_inputs[0].data.numpy()
```

### Comments on the visualization

There are a few good things happening here. As the agent layer remains the same across all observations, we would expect that the agent has **learned to ignore the `agent` layer**. The EB saliency map seems to confirm this suspicion. Looking at the other layers, the agent appears to be **attending to Towers 1 and 4 in particular**. Upon looking at some other inputs, I noticed that this was part of a larger pattern; the EB saliency would generally highlight **two towers out of the four**.

What is special about Towers 1 and 4 (versus 2 and 3)? There are many possible reasons, but I will take a guess. I noticed that **Tower 4 has the largest health** of the four towers. In looking over several other examples, I noticed that EB generally focused saliency upon the tower with the largest health value (regardless of other attributes). Remember, though, that the hit points of a large tower are calculated as `health*1.5` whereas the hit points for a small one are `heath*0.5`. The `friend` versus `enemy` layers matter as well. In other words, **choosing the tower with max health is not an optimal policy**. So, how does the agent choose Tower 1 as a comparison? It may be that Tower 1 is the only enemy of the remaining options. Generally speaking, the agent will get a higher reward for choosing to attack an enemy than for choosing to attack a friend, so it would make sense if the policy has **learned to give enemy towers preference**.

**Why does EB highlight just two towers?** I wonder if this is an **unintended consequence of including redundant channels** for `tower size` (`large` and `small`) and `tower team` (`friend` and `enemy`). In both cases, we could have provided only one of those channels and the agent would have been able to infer that, say, a tower is _large_ if it is not _small_. Anyways, I noticed that Tower 1 has its EB saliency concentrated in the `small` and `enemy` channels whereas Tower 4 has its saliency concentrated in the `large` and `friend` channels. More generally, the policy _always_ seems to compare two of the four towers and one of these towers has its saliency concentrated in `small` and `enemy` whereas the other has its saliency concentrated in `large` and `friend`. So we see that the agent may be **exploiting the redundancy of the channels** to obtain information about one tower from one set of channels and information about the other tower from another set of channels.

The crucial insight here is that the policy's **behavior has been shaped by the _format_ of the input** information we gave it.

## Let's perturb only the health of the agent

This will help us answer the question _"At what health values (for itself) will the agent choose to attack Tower 1?"_

```python
tower1_health = obs[0,:l,l:2*l].sum()*0.5 # multiply by 0.5 because it's a small tower

healths = np.linspace(0,1,100)
probs = np.zeros_like(healths)
rew = np.zeros_like(healths)
for i, v in enumerate(healths):
    pobs = obs.copy()
    pobs[0,5,5] = v
    _, dist = model(state_to_torch(pobs))
    probs[i] = dist.data.view(-1)[0]
    
    env.obs[0,5,5] = v
    rew[i] = env.step(0)[1]

plt.figure(figsize=[12,4])
plt.subplot(1,2,1)
plt.plot(healths, rew, 'g-', label='reward')
plt.title('reward curve')
plt.ylim(-1.2,1.2)
plt.xlabel('agent health')
plt.ylabel('reward')


plt.subplot(1,2,2)
plt.plot(healths, probs, 'b-', label='p(agent choose tower 1)')
plt.plot([tower1_health]*2, [0,1], 'r-', label='tower 1 health')
plt.legend()
plt.title('decision threshold')
plt.ylim(0,1.2)
plt.xlabel('agent health')
plt.ylabel('probability of attack')

plt.show()
```
